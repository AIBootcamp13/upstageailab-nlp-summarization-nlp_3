```python
import torch
from transformers import pipeline
from sentence_transformers import SentenceTransformer
import re

import numpy as np

class DialogueParaphraser:
    """
    í™”ì ì •ë³´ë¥¼ ì™„ë²½í•˜ê²Œ ë³´ì¡´í•˜ë©´ì„œ ëŒ€í™”ë¥¼ íŒ¨ëŸ¬í”„ë ˆì´ì§•í•˜ëŠ” í´ë˜ìŠ¤
    """
    def __init__(self):
        self.paraphraser = None
        self.model_name = "psyche/KoT5-paraphrase-generation" # ì‚¬ìš©í•˜ì‹œë ¤ëŠ” ëª¨ë¸
        
        try:
            print(f"'{self.model_name}' ëª¨ë¸ ë¡œë”© ì‹œë„ ì¤‘...")
            self.paraphraser = pipeline(
                "text2text-generation",
                model=self.model_name,
                device=0 if torch.cuda.is_available() else -1
            )
            print(f"âœ… '{self.model_name}' ëª¨ë¸ ë¡œë”© ì„±ê³µ!")

        except Exception as e:
            print(f"âŒ ëª¨ë¸ ë¡œë”© ì‹¤íŒ¨: {e}")
            # ë¡œë”© ì‹¤íŒ¨ ì‹œ ëŒ€ì•ˆ ëª¨ë¸ ì‹œë„
            self.model_name = "gogamza/kobart-summarization"
            print(f"ëŒ€ì•ˆ ëª¨ë¸ '{self.model_name}' ë¡œë”© ì‹œë„ ì¤‘...")
            self.paraphraser = pipeline(
                "summarization",
                model=self.model_name,
                device=0 if torch.cuda.is_available() else -1
            )
            print(f"âœ… ëŒ€ì•ˆ ëª¨ë¸ '{self.model_name}' ë¡œë”© ì„±ê³µ!")


    def paraphrase_dialogue(self, dialogue: str) -> str:
        """
        ëŒ€í™” êµ¬ì¡°ë¥¼ ë³´ì¡´í•˜ë©° ì•ˆì „í•˜ê²Œ íŒ¨ëŸ¬í”„ë ˆì´ì§•í•©ë‹ˆë‹¤.
        """
        # 1. ì •ê·œí‘œí˜„ì‹ì„ ì‚¬ìš©í•´ í™”ì íƒœê·¸ì™€ ë°œí™” ë‚´ìš©ì„ ì •í™•íˆ ë¶„ë¦¬í•©ë‹ˆë‹¤.
        speaker_pattern = r'(#(?:Person|ê°œì¸)\d+#:)'
        parts = re.split(speaker_pattern, dialogue)
        
        segments = []
        # parts ë¦¬ìŠ¤íŠ¸ëŠ” [ '', '#Person1#:', ' ë°œí™” ë‚´ìš©1 ', '#Person2#:', ' ë°œí™” ë‚´ìš©2' ] í˜•íƒœë¡œ ë‚˜ì˜µë‹ˆë‹¤.
        for i in range(1, len(parts), 2):
            speaker_tag = parts[i]
            utterance = parts[i+1].strip()
            if utterance:
                segments.append((speaker_tag, utterance))

        paraphrased_segments = []
        for speaker_tag, utterance in segments:
            try:
                # 2. ìˆœìˆ˜í•œ ë°œí™” ë‚´ìš©(utterance)ë§Œ ëª¨ë¸ì— ì „ë‹¬í•©ë‹ˆë‹¤.
                #    ëª¨ë¸ì´ 'summarization'ì¼ ê²½ìš°ë¥¼ ëŒ€ë¹„í•´ íƒœìŠ¤í¬ ì´ë¦„ì„ í™•ì¸í•©ë‹ˆë‹¤.
                
                # text2text-generation ëª¨ë¸ ì‚¬ìš©
                result = self.paraphraser(
                    utterance,
                    max_new_tokens=512,
                    do_sample=True,
                    temperature=0.8,          # ë” ë†’ì€ ì˜¨ë„ë¡œ ì°½ì˜ì„± ì¦ê°€
                    top_k=50,                 # top_k ê°’ì„ ì„¤ì •í•˜ì—¬ ë‹¤ì–‘ì„± í™•ë³´
                    repetition_penalty=1.2,   # ë°˜ë³µì„ ì¤„ì—¬ ìƒˆë¡œìš´ í‘œí˜„ ìœ ë„
                    num_beams=1,              # ë¹” ì„œì¹˜ ëŒ€ì‹  ìƒ˜í”Œë§ì— ì§‘ì¤‘
                    num_return_sequences=1
                )
                paraphrased_utterance = result[0]['generated_text']

                # 3. íŒ¨ëŸ¬í”„ë ˆì´ì§•ëœ ê²°ê³¼ì— ì›ë˜ì˜ í™”ì íƒœê·¸ë¥¼ ë‹¤ì‹œ ë¶™ì—¬ì¤ë‹ˆë‹¤.
                paraphrased_segments.append(f"{speaker_tag} {paraphrased_utterance}")

            except Exception:
                # íŒ¨ëŸ¬í”„ë ˆì´ì§• ì‹¤íŒ¨ ì‹œ, ì•ˆì „í•˜ê²Œ ì›ë³¸ì„ ì‚¬ìš©í•©ë‹ˆë‹¤.
                paraphrased_segments.append(f"{speaker_tag} {utterance}")
        
        # 4. ê°œí–‰ë¬¸ì(\n)ë¡œ ê° ë°œí™”ë¥¼ ì—°ê²°í•˜ì—¬ ìµœì¢… ëŒ€í™”ë¥¼ ë§Œë“­ë‹ˆë‹¤.
        return "\n".join(paraphrased_segments)
    
    def calculate_semantic_similarity(self, text1: str, text2: str) -> float:
        """ì˜ë¯¸ ê¸°ë°˜ ìœ ì‚¬ë„ ê³„ì‚° (ê¸°ì¡´ Jaccard ëŒ€ì‹ )"""
        
        # ëª¨ë¸ì´ ì—†ë‹¤ë©´ ì´ˆê¸°í™”
        if not hasattr(self, 'similarity_model'):
            print("ìœ ì‚¬ë„ ì¸¡ì • ëª¨ë¸ ë¡œë”© (ìµœì´ˆ 1íšŒ)...")
            self.similarity_model = SentenceTransformer('jhgan/ko-sroberta-multitask')
        
        # ë¬¸ì¥ì„ ì„ë² ë”© ë²¡í„°ë¡œ ë³€í™˜
        embeddings = self.similarity_model.encode([text1, text2])
        
        # ì½”ì‚¬ì¸ ìœ ì‚¬ë„ ê³„ì‚°
        vec1 = embeddings[0]
        vec2 = embeddings[1]
        
        similarity = np.dot(vec1, vec2) / (np.linalg.norm(vec1) * np.linalg.norm(vec2))
        
        return similarity

try:
    # 1. í´ë˜ìŠ¤ ì¸ìŠ¤í„´ìŠ¤ ìƒì„±
    final_paraphraser = DialogueParaphraser()

    # 2. ë¬¸ì œê°€ ë°œìƒí–ˆë˜ ì›ë³¸ ëŒ€í™”
    test_dialogue = """#Person1#: Mr. King, ì œê°€ ë„ì™€ë“œë¦´ ì¼ ìˆë‚˜ìš”?
#Person2#: ë‚´ì¼ ì•„ì¹¨ì— ìƒí•˜ì´ë¡œ ë– ë‚©ë‹ˆë‹¤. ê³„ì‚°ì€ ì–¸ì œ í•´ì•¼ í•˜ë‚˜ìš”?
#Person1#: ì˜¤ëŠ˜ ì €ë…ì— ê³„ì‚°í•˜ì…”ë„ ê´œì°®ìœ¼ì‹ ê°€ìš”?
#Person2#: ê´œì°®ìŠµë‹ˆë‹¤. ëˆì€ ì–´ë”” ê°€ì„œ ë‚´ì•¼ í•˜ë‚˜ìš”?
#Person1#: ì €í¬ í˜¸í…” ë¡œë¹„ì— ìˆëŠ” ìˆ˜ë‚© ë°ìŠ¤í¬ì—ì„œ í•˜ì‹œë©´ ë©ë‹ˆë‹¤.
#Person2#: ì €ë…ì‹ì‚¬ ì „ê¹Œì§€ ê³„ì‚°ì„œ ì¤€ë¹„í•´ ì£¼ì‹œê² ì–´ìš”?
#Person1#: ì•Œê² ìŠµë‹ˆë‹¤. ë™í–‰í•˜ì‹  ë¶„ë„ ìƒí•˜ì´ë¡œ í•¨ê»˜ ê°€ì‹œë‚˜ìš”?
#Person2#: ì•„ë‹ˆìš”, ê·¸ëŠ” ì´ê³³ì— ì´í‹€ ë” ë¨¸ë¬¼ ì˜ˆì •ì…ë‹ˆë‹¤. ë°©ì€ ê³„ì† ì‚¬ìš©í•  ìˆ˜ ìˆê²Œ í•´ì£¼ì„¸ìš”.
#Person1#: í•œ ì¥ìœ¼ë¡œ ê³„ì‚°ì„œë¥¼ ì¤€ë¹„í• ê¹Œìš”, ì•„ë‹ˆë©´ ë‘ ê°œë¡œ ë‚˜ëˆŒê¹Œìš”?
#Person2#: ë‘ ê°œë¡œ ë‚˜ëˆ  ì£¼ì„¸ìš”.
#Person1#: ì•Œê² ìŠµë‹ˆë‹¤. ì €ë…ì— ê³„ì‚°ì„œ ì¤€ë¹„í•´ ë“œë¦¬ê² ìŠµë‹ˆë‹¤.
#Person2#: ì—¬ì„¯ ì‹œì¯¤ì— ê³„ì‚°ì„œ ë°›ì„ ìˆ˜ ìˆì„ê¹Œìš”?
#Person1#: ë¬¸ì œì—†ìŠµë‹ˆë‹¤. ê·¸ëŸ°ë° ì €í¬ í˜¸í…”ì— ëŒ€í•´ ì˜ê²¬ì´ ìˆìœ¼ì‹ ê°€ìš”?
#Person2#: ì—¬ê¸°ì„œ í¸ì•ˆí•˜ê²Œ ì§€ëƒˆìŠµë‹ˆë‹¤. ì„œë¹„ìŠ¤ê°€ í›Œë¥­í•˜ë„¤ìš”. ì•„ì£¼ ë§Œì¡±í•©ë‹ˆë‹¤.
#Person1#: ì¹­ì°¬ ê°ì‚¬í•©ë‹ˆë‹¤. ì–¸ì œë“  ë‹¤ì‹œ ì˜¤ì‹­ì‹œì˜¤.
#Person2#: ë„¤, ê·¸ëŸ´ê²Œìš”."""

    # 3. íŒ¨ëŸ¬í”„ë ˆì´ì§• ì‹¤í–‰
    paraphrased_text = final_paraphraser.paraphrase_dialogue(test_dialogue)

    # 4. ê²°ê³¼ í™•ì¸
    print("\n" + "="*20 + " ìµœì¢… ê²°ê³¼ " + "="*20)
    print("### ì›ë³¸ ëŒ€í™” ###")
    print(test_dialogue)
    print("\n### íŒ¨ëŸ¬í”„ë ˆì´ì§• ê²°ê³¼ (êµ¬ì¡° ë³´ì¡´ë¨) ###")
    print(paraphrased_text)
    print("="*53)
    print("similarity:"+ str(final_paraphraser.calculate_semantic_similarity(test_dialogue, paraphrased_text)))    
except Exception as e:
    print(f"ìµœì¢… ì‹¤í–‰ ì¤‘ ì˜¤ë¥˜ ë°œìƒ: {e}")
```

```python
def context_aware_removal(text):
    """ë¬¸ë§¥ì„ ê³ ë ¤í•œ íƒœê·¸ ì²˜ë¦¬"""
    
    # HTML íƒœê·¸ íŠ¹ì§•: ì†Œë¬¸ì, ì•Œë ¤ì§„ íƒœê·¸ëª…
    html_pattern = r'<(div|p|span|a|img|br|hr|h[1-6]|ul|ol|li|table|tr|td|strong|b|em|i)[^>]*>'
    text = re.sub(html_pattern, '', text, flags=re.IGNORECASE)
    
    # ë‹«ëŠ” íƒœê·¸
    text = re.sub(r'</(div|p|span|a|h[1-6]|ul|ol|li|table|tr|td|strong|b|em|i)>', '', text, flags=re.IGNORECASE)
    
    # í™”ì íƒœê·¸ëŠ” ê·¸ëŒ€ë¡œ ë‘ê±°ë‚˜ ë³€í™˜
    # <ì´ë¦„> íŒ¨í„´ì„ ì´ë¦„: ìœ¼ë¡œ ë³€í™˜
    text = re.sub(r'<([A-Za-zê°€-í£\s]+)>', r'\1:', text)
    
    return text


test_cases = [
    "<The Witch> ë§ˆë²•ì„ ì‚¬ìš©í•©ë‹ˆë‹¤",
    "<Player1> ì•ˆë…•í•˜ì„¸ìš”",
    "<div class='content'>HTML ë‚´ìš©</div>",
    "<p>ë¬¸ë‹¨ì…ë‹ˆë‹¤</p>", 
    "<a href='link'>ë§í¬</a>",
    "<System> ì‹œìŠ¤í…œ ë©”ì‹œì§€",
    "<br>",
    "<img src='image.jpg'>",
    "<span>í…ìŠ¤íŠ¸</span>"
]

print("ì»¨í…ìŠ¤íŠ¸ ê¸°ë°˜ ì²˜ë¦¬ ê²°ê³¼:")
for text in test_cases:
    result = context_aware_removal(text)
    print(f"'{text}' â†’ '{result}'")
```
